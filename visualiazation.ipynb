```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.histplot(iris , x ='sepal_width' , bins=10 , kde = True)
```

```python
fig , ax = plt.subplots(figsize = (5,3))
sns.histplot(iris , x ='sepal_length' , bins=10 , kde = True  , ax=ax)
plt.savefig('sepal_length_hist.png' , dpi=300)
```

```python
fig , ax = plt.subplots(figsize = (5,3))
sns.histplot(iris , x ='petal_width' , bins=10 , kde = True)
plt.savefig('sepal_width_hist.png' , dpi=300)
```

```python
fig , ax = plt.subplots(figsize = (5,3))
sns.histplot(iris , x ='petal_length' , bins=10 , kde = True ,ax=ax)
plt.savefig('petal_length_hist.png' , dpi=500)
```

```python
fig , ax = plt.subplots(figsize = (5,3))
sns.histplot(iris , x ='petal_width' , bins=10 , kde = True)
plt.savefig('petal_width_hist.png' , dpi=300)
```

```python
sns.color_palette()
```

```python
sns.set_palette('rainbow')
fig,ax = plt.subplots(figsize=(20,5) , ncols=4)

sns.histplot(iris , x ='sepal_width' , bins=10 , kde = True  , ax=ax[0], color= 'red')
sns.histplot(iris , x ='petal_width' , bins=10 , kde = True  , ax=ax[1] , color = 'green')
sns.histplot(iris , x ='sepal_length' , bins=10 , kde = True  , ax=ax[2] , color='red')
sns.histplot(iris , x ='petal_length' , bins=10 , kde = True  , ax=ax[3] , color='green')
```

```python
sns.set_palette('rainbow')
fig,ax = plt.subplots(figsize=(20,7), ncols= 2, nrows=2)

sns.histplot(iris , x ='sepal_width' , bins=10 , kde = True  , ax=ax[0,0], color= 'red')
sns.histplot(iris , x ='petal_width' , bins=10 , kde = True  , ax=ax[0,1] , color = 'green')
sns.histplot(iris , x ='sepal_length' , bins=10 , kde = True  , ax=ax[1,0] , color='red')
sns.histplot(iris , x ='petal_length' , bins=10 , kde = True  , ax=ax[1,1] , color='green')
```
