```python
def scrape_product(query,page=1):
    webpage = Scraper(f'https://www.amazon.in/s?k={query}&page={page}')
    print(webpage.soup.title)
    #important

    target = Tag ('div' , cls = 's-main-slot s-result-list s-search-results sg-row')
    items = Tag ('div' , cls= 's-result-item')    # repeating products
    title = Tag ('h2')
    price = Tag ('span' , cls = 'a-price-whole')
    rating = Tag ('i' , cls = 'a-icon a-icon-star-small a-star-small-3-5 aok-align-bottom')

    return webpage.get_all (target, items, title=title, price=price, rating=rating )
    
    
```

```python
all_data = []
query = 'laptop'
for i in range(1 , 15):
    all_data.extend(scrape_product(query,i))

    print( f'Page {i} done')

all_data
```

```python
import pandas as pd
df = pd.DataFrame(all_data).dropna(how='all')
df.to_csv('laptops.csv' , index= False)
```

## Flipkart Mobile Scraper 

```python
from dputils.scrape import Scraper , Tag


```

```python
query = 'phones'
page = 1
url = f'https://www.flipkart.com/search?q={query}&page={page}'
scraper = Scraper(url)
scraper.soup.title
```

```python
target = Tag ('div' , cls = '_1YokD2 _3Mn1Gg"')
items = Tag ('div' , cls= '_1AtVbE col-12-12')    # repeating products
title = Tag (cls = '_4rR01T')
price = Tag (cls = '_30jeq3 _1_WHN1')
rating = Tag ('span' , cls = '_2_R_DZ')
out = scraper.get_all(target,items,name = title ,price = price , rr = rating)

print(out)
```
